```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import os
import lightkurve as lk
import astropy.units as u
import numpy as np
```

```python
Target_name = str("TIC")+"152825521"
lc = lk.search_lightcurvefile(Target_name, mission="TESS").download_all().SAP_FLUX
lightcurve = lc.stitch()
lc_norm = lightcurve#.normalize()
lc_norm.scatter()

lc_norm.to_csv('TIC152825521_SAP_normalized_flux')

#pg = lc_norm.to_periodogram(minimum_period=0.2*u.day, maximum_period=20*u.day, oversample_factor=25)
#print (pg.period_at_max_power)
#period_2 = 0.60675691674
#lc_fig = lc_norm.fold(period_2)
#lc_fig.scatter()
#lc_fig_df = lc_fig.to_pandas()
#lc_norm_df = lc_norm.to_pandas()
#lc_fig_df.to_csv(Target_name+str("sector32")+str("_lc_phasefolded"))
#lc_norm_df.to_csv(Target_name+str("sector32")+str("_SAP_lc"))
#pg_norm = lc_fig.to_periodogram(minimum_period=0.001*u.day, maximum_period=0.4*u.day, oversample_factor=25)
#pg_norm.plot()
```

```python
pg = lc.to_periodogram(minimum_period=0.2*u.day, maximum_period=20*u.day, oversample_factor=25)
period_2 = 2*(pg.period_at_max_power)
lc_fig = lc_norm.fold(period_2)
lc_fig_df = lc_fig.to_pandas()
lc_norm_df = lc_norm.to_pandas()
lc_fig_df.to_csv(Target_name+str("_lc_phasefolded"))
lc_norm_df.to_csv(Target_name+str("_PDCSAP_lc"))
makedir_command=str("mkdir ")+ str(Target_name)
os.system(makedir_command)
#______________________________________________________________________________________
makecsv_lc_command=str("mv ")+ str(Target_name)+str("_lc_phasefolded ")+str(Target_name)+str("/")
os.system(makecsv_lc_command)
#______________________________________________________________________________________
make_lc_command=str("mv ")+ str(Target_name)+str("_PDCSAP_lc ")+str(Target_name)+str("/")
os.system(make_lc_command)
#______________________________________________________________________________________
#make_SAP_lc_command=str("mv ")+ str(Target_name)+str("_SAP_lc ")+str(Target_name)+str("/")
#os.system(make_SAP_lc_command)
#--------------------------------------------------------------------------------------
Title = Target_name + str(" (approxperiodindays_")+ str(period_2)+ str(")")
plot = lc_fig_df.plot(x="time",y="flux", title=Title, marker = ".", markersize = 0.5)
plot.set_xlabel("Phase")
plot.set_ylabel("Normalized flux")
plot.get_figure().savefig(Target_name, dpi=500)
move_command= str("mv ")+str(Target_name)+str(".png ")+str(Target_name)+str("/")
os.system(move_command)
```

```python
from lightkurve import search_targetpixelfile

tpf1 = search_targetpixelfile("TIC 231224520", campaign=111).download()
tpf1
tpf2 = search_targetpixelfile("TIC 231224520", campaign=112).download()
tpf2
```

```python
lc1 = tpf1.to_lightcurve(aperture_mask=tpf1.pipeline_mask)
lc1.remove_nans()
lc1.remove_outliers()
lc1_norm = lc1.normalize()
lc1.scatter()

lc2 = tpf2.to_lightcurve(aperture_mask=tpf2.pipeline_mask)
lc2.remove_nans()
lc2.remove_outliers()
lc2_norm = lc2.normalize()
lc2.scatter()

lc_all = lk.LightCurve.append(lc1_norm, lc2_norm)
```
